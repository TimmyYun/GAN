```python
import os
import cv2 
from PIL import Image 

image_folder = '/raid/timur_unaspekov/datasets/voxceleb2/dev/part1'
video_name = '/raid/timur_unaspekov/datasets/voxceleb2/metadata/videos/part1.avi'

images = []

mean_height = 0
mean_width = 0

for root, dirs, files in os.walk(image_folder):
    for file in files:    
        if file == '00001.jpg':
            images.append(os.path.join(root, file))
            
num_of_images = len(images)
print(num_of_images)

for image in images:
    im = Image.open(image)
    width, height = im.size
    mean_width += width
    mean_height += height

mean_width = int(mean_width / num_of_images)
mean_height = int(mean_height / num_of_images)

print(mean_height, mean_width)

count = 1
for file in images:
        im = Image.open(file) 

        width, height = im.size   
  
        imResize = im.resize((mean_width, mean_height), Image.ANTIALIAS) 
        imResize.save(f'/raid/timur_unaspekov/datasets/voxceleb2/metadata/resized_images/part1/{count}', 'JPEG', quality = 95)
        print(f"{count}/{num_of_images} images are resized") 
        count += 1

        
# frame = cv2.imread(images[0])

# height, width, layers = frame.shape  

# video = cv2.VideoWriter(video_name, 0, 1, (width, height)) 

# for image in images: 
#     video.write(cv2.imread(os.path.join(image_folder, image))) 

# cv2.destroyAllWindows() 
# video.release() 
```

```python
resized_images = []

for root, dirs, files in os.walk('/raid/timur_unaspekov/datasets/voxceleb2/metadata/resized_images/part1/'):
    for file in files: 
        resized_images.append(os.path.join(root, file))

frame = cv2.imread(resized_images[0])

height, width, layers = frame.shape  

video = cv2.VideoWriter(video_name, 0, 1, (width, height)) 

for image in resized_images: 
    video.write(cv2.imread(image)) 

cv2.destroyAllWindows() 
video.release() 
```
